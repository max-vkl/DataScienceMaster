```python
def calculate_MAP(csv_file, lookup_table):
    # Read the CSV file
    df = pd.read_csv(csv_file)
    
    # Split the prediction column by space and convert it to a list of integers
    df['prediction'] = df['prediction'].str.split().apply(lambda x: list(map(int, x)))
    
    # Merge the dataframe with the lookup table
    merged_df = df.merge(lookup_table, on='user_id', how='left')
    
    # Calculate the number of relevant movies for each user
    merged_df['num_relevant_movies'] = merged_df.apply(lambda row: len(set(row['prediction']).intersection(set(row['movie_ids']))), axis=1)
    
    # Calculate the Average Precision for each user
    merged_df['average_precision'] = merged_df['num_relevant_movies'] / merged_df['prediction'].apply(lambda x: len(x))
    
    # Calculate the Mean Average Precision (MAP)
    MAP = merged_df['average_precision'].mean()
    
    return MAP/2
```
